```python
import numpy as np
import scipy.spatial.distance as sp
from heapq import heappush, heappop

result = []
with open('bigmaze.txt') as maze:
    for line in maze:
        result.append(list(line.strip()))
for n in range(len(result)):
    print("".join(result[n]))
# Method to read txt file adapted from the following link
# https://stackoverflow.com/questions/18448847/import-txt-file-and-having-each-line-as-a-list

result_0 = result[1:-1]
result_1 = []
for n in range(len(result_0)):
    result_1.append(result_0[n][1:-1])
# remove all walls

for m in range(len(result_1)):
    for n in range(len(result_1[m])):
        if result_1[m][n] == "%":
            result_1[m][n] = 1
            # convert to 1 if wall
        elif result_1[m][n] == ".":
            print ("The path ends at ({},{})".format(m+1,n+1))
            end_x = m
            end_y = n
            # print ending point
            result_1[m][n] = 0
            # convert to 0 if ending point
        elif result_1[m][n] == "P":
            start_x = m
            start_y = n
            print ("The path starts at ({},{})".format(m+1,n+1))
            # print starting point
            result_1[m][n]= 0
            # convert to 0 if starting point
        else: result_1[m][n] = 0
            # convert to 0 if vacant

result_2 = np.array(result_1)

def astar(maze, start, end):                            # pseudocode at http://web.mit.edu/eranki/www/tutorials/search/
    open_list = []                                      # initialize the open list
    closed_list = []                                    # initialize the closed list
    f = {}                                              # recording fn of the nodes 
    f[start] = sp.cityblock(list(start),list(end))      # f(start) = h(start)
    open_list.append((f[start], start))                 # put the starting node on the open list
    parent = {}                                         # recording the matching parent of a successor
    node = 0                                            # expanded node counter
    
    while (len(open_list) != 0):
        node += 1
        q = heappop(open_list)[1]
        # find the node with the least f on the open list, call it "q",  pop q off the open list

        for successor in [(q[0]+1, q[1]+0), (q[0]-1, q[1]+0), (q[0]+0, q[1]+1), (q[0]+0, q[1]-1)]:
        # loop through four successors in NSEW directions
            
            if successor == end:
                path = []
                path.append(end)
                while q in parent:
                    path.append(q)
                    q = parent[q]
                print ("The path cost (# of steps): {}".format(len(path)))
                print ("Number of nodes expanded: {}".format(node+1))
                return path
            # if successor is the goal, stop the search
            
            if (0 <= successor[0] < maze.shape[0]) and (0 <= successor[1] < maze.shape[1]) and (maze[successor[0]][successor[1]]) == 1:
                continue
            # successor in range but is wall (unreachable), stop the search
            
            if (not 0 <= successor[0] < maze.shape[0]) or (not 0 <= successor[1] < maze.shape[1]):
                continue
            # successor out of range, stop the search
                
            fvalue = f[q] + sp.cityblock(list(successor),list(end)) + 1
            # f score equals (old f score) plus (the heuristic of the new node) plus (one)
            
            if (successor in closed_list) or ((successor in [i[1] for i in open_list]) and (fvalue >= f[successor])):
                continue
            # if (successor in the closed list) OR (successor in the open list AND has larger f value), stop the search 

            if successor not in [i[1] for i in open_list]:
                parent[successor] = q
                f[successor] = fvalue
                heappush(open_list, (f[successor], successor))

        closed_list.append(q)
        # push q on the closed list
        
    return None

path = (astar(result_2, (start_x,start_y), (end_x,end_y)))

for n in path:
    result[n[0]+1][n[1]+1] = "."

result[start_x+1][start_y+1] = "P"
for n in range(len(result)):
    print("".join(result[n]))
```
